1 Read the first 10 objects from the gamedata.json file into the DataFrame. Then read the last five objects from the same file into another DataFrame. Combine these two into one DataFrame.

```python
import pandas as pd

gamedata = pd.read_json('data_files/gamedata.json')

gamedata = pd.concat([gamedata.head(10),gamedata.tail()])
gamedata.reset_index(inplace=True,drop=True)
gamedata
```

2 Set *steamRatingPercent* to NaN for all rows having *steamRatingCount* below 1000.

```python
import numpy as np

gamedata.loc[gamedata['steamRatingCount'] < 1000, 'steamRatingPercent'] = np.nan
gamedata
```

3 Fill all NaN values in *steamRatingPercent* column by increasing the *metacriticScore* from the same row with 7 %.

```python
gamedata['steamRatingPercent'].fillna(inplace=True,value=gamedata['metacriticScore']*1.07)
gamedata
```

4 Group the data by *steamRatingText* column so that averages of the following columns will be calculated for each text value: *steamRatingPercent*, *metacriticScore*, *normalPrice* and *salePrce*. Do not show any other columns in the resulting DataFrame.

```python
gamedata.groupby('steamRatingText')[['steamRatingPercent','metacriticScore','normalPrice','salePrice']].mean().round(2)
```
