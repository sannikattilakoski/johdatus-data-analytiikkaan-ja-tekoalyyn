1 Use gamedata.json for this task. This file contains information of games sold through Steam. Parse out the following information from the data:
- TOP 3 highest metacritic score. Present results using the following format: *Title* has metacritic score of *Score* (for example)
- Games with price discount being 90 % or more. Present results using the following format: *Title* | Discount: *Savings* (for example Metal Gear Solid V: Ground Zeroes | Discount: 90.090090)
- Games having metacritic score higher than steam score. Present results using the following format: *Title* has metacritic score of *MetacriticScore* and steam score of *SteamRatingPercent*

```python
import json

with open("data_files/gamedata.json") as json_data:
    games_json = json.load(json_data)  
```

```python
#TOP 3 highest metacritic score:#
x = sorted(games_json,key=lambda i:i['metacriticScore'])[::-1][:3]

for i in x:
    print(f"{i['title']} has metacritic score of {i['metacriticScore']}")     
```

```python
#Games with price discount being 90 % or more
biggest_discount = []

for i in games_json:
    if float(i["savings"]) >= 90:
        biggest_discount.append(i)

for i in biggest_discount:
    print(f"{i['title']} | Discount: {i['savings']}") 
```

```python
#Games having metacritic score higher than steam score
metacriticscores = []

for i in games_json:
    if int(i['metacriticScore']) > int(i['steamRatingPercent']):
        metacriticscores.append(i)

for i in metacriticscores:
    print(f"{i['title']} has metacritic score of {i['metacriticScore']} and steam score of {i['steamRatingPercent']}")
```

2 Use earthquakes.csv for this task. This file contains information about earthquakes recorded between 1965 and 2016. Earthquake magnitude value describes how strong the earthquake is. Magnitude information can be categorized like presented in the table below (*Source: http://www.geo.mtu.edu/UPSeis/magnitude.html*).

| Magnitude      | Class | Effects |
|----------------|-------|---------|
| 2.4 or less    | Minor | Usually not felt, but can be recorded by seismograph. |
| 2.5 to 5.4     | Light | Often felt, but only causes minor damage. |
| 5.5 to 6.0     | Moderate | Slight damage to buildings and other structures. |
| 6.1 to 6.9     | Strong | May cause a lot of damage in very populated areas. |
| 7.0 to 7.9     | Major | Major earthquake. Serious damage. |
| 8.0 or greater | Great | Great earthquake. Can totally destroy communities near the epicenter. |

Count how many earthquakes have occurred in each class.

<b style="color:red;">Notice:</b> The first value has been modified to be 2.4 or less compared to the original source (has been 2.5 or less).

```python
import csv

with open("data_files/earthquakes.csv") as file:
    earthquakes_csv = csv.reader(file)
    earthquake_data = []
    for i in earthquakes_csv:
        earthquake_data.append(i)

#titles as keys in dictionary        
titles = earthquake_data[0]

earthquakes = earthquake_data[1:]

earthquake_collection = []

for item in earthquakes:
    obj = {}
    for j in range(len(item)):
        obj[titles[j]] = item[j]
    earthquake_collection.append(obj)
```

```python
minor = [] 
light= []
moderate = []
strong = []
major = []
great = []

for e in earthquake_collection:
    magnitude = float(e["Magnitude"])
    if magnitude <= 2.4:
        minor.append(magnitude)
    elif magnitude > 2.5 and magnitude < 5.4:
        light.append(magnitude)
    elif magnitude > 5.5 and magnitude < 6.0:
        moderate.append(magnitude)
    elif magnitude > 6.1 and magnitude < 6.9:
        strong.append(magnitude)
    elif magnitude > 7.0 and magnitude < 7.9:
        major.append(magnitude)
    else:
        great.append(magnitude)
       
print(f"Minor: {len(minor)}, Light: {len(light)}, Moderate: {len(moderate)}, Strong: {len(strong)}, Major: {len(major)}, Great: {len(great)}")
```

3 Use netflix_titles.xml for this task. This file contains information about Netflix movies and TV shows. **Important:** Movies have duration presented in minutes while TV shows have duration presented in amount of seasons! Parse out the following information from the data:
- Movies released in 2017
- TV show and movie amount (present both counts in separate lines)
- Movies with a length between 15 and 20 minutes (values 15 and 20 included)

```python
import xml.etree.ElementTree as e

tree = e.parse("data_files/netflix_titles.xml")
root = tree.getroot()

movie_list = []
for i in root:
    obj = {}
    for j in i:
        obj[j.tag] = j.text
    movie_list.append(obj)
```

```python
#Movies released in 2017
for i in movie_list:
    if int(i["release_year"]) == 2017 and i["type"] == "Movie":
        print(i["title"])
```

```python
#TV show and movie amount
tv_shows = 0
movies = 0

for i in movie_list:
    if i["type"] == "TV Show":
        tv_shows+=1
    else:
        movies+=1
        
print(f"TV Shows: {tv_shows}")
print(f"Movies: {movies}")
```

```python
# Movies with a length between 15 and 20 minutes (values 15 and 20 included)        
for i in movie_list:
     if i["type"] == "Movie":
        i['duration'] = i['duration'].replace('min', '')
        if int(i['duration']) >= 15 and int(i['duration']) <= 20:
            print(i['title'])
```

```python
tv_shows = 0
movies = 0
for i in movie_list:
    if i["type"] == "TV Show":
        tv_shows+=1
    else:
        movies+=1
print(f"TV Shows: {tv_shows}")
print(f"Movies: {movies}")
```

4 Use the following Rest API for this task: https://tie.digitraffic.fi/api/v1/data/weather-data. Calculate the average for air temperature (ILMA) and humidity (ILMAN_KOSTEUS) values using two decimals.

```python
import requests

url = "https://tie.digitraffic.fi/api/v1/data/weather-data"

req = requests.get(url=url)

req.status_code
```

```python
weather_data = req.json()
#weather_data

air_temperature = []
humidity= []

for i in weather_data['weatherStations']:
    for j in i['sensorValues']:
        if j['name'] == 'ILMA':
            air_temperature.append(float(j['sensorValue']))
        elif j['name'] == 'ILMAN_KOSTEUS':
             humidity.append(float(j['sensorValue']))

avg_air = round(sum(air_temperature)/len(air_temperature),2)
avg_humidity = round(sum(humidity)/len(humidity),2)

print(f"Average air temperature is {avg_air} and average humidity is {avg_humidity}")
```
